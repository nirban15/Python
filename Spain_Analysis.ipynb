```python
# In the first case, the inflation rate is lower than the GDP growth rate. This means that the purchasing power of the currency is increasing. This is a good sign for the economy, 
# as it means that people are able to buy more goods and services with their money.
# In the second case, the inflation rate is higher than the GDP growth rate. This means that the purchasing power of the currency is decreasing. This is a bad sign for the economy, 
# as it means that people are able to buy fewer goods and services with their money.
```

```python
y=spainmacro_ind['GDP growth (annual %)']
```

```python
x=spainmacro_ind[['Population growth (annual %)','CO2 emissions (metric tons per capita)',
                 'Industry (including construction), value added (% of GDP)',
                 'Exports of goods and services (% of GDP)','Imports of goods and services (% of GDP)',
                 'Gross capital formation (% of GDP)','Inflation, consumer prices (annual %)']]
```

```python
x.corr()
# correlation value of the independent variable is not more than 0.95. 
# thus we can proceeds to built model as there is no multicollinearity
```

```python
spainmacro_ind.info()
```

```python
# Multiple linear regression
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
regmodel=reg.fit(x,y)
```

```python
regmodel.score(x,y) 
# Since R square value is 0.5807106291099586. So, the independent variables
# can explain 69% of the changes in independent variable i.e. GDP growth.
```

```python
regpredict=regmodel.predict(x)
```

```python
regresidual=regmodel.predict(x)
```

```python
regresidual=y-regpredict
```

```python
np.sqrt(np.mean(regresidual**2)) 
# RMSE takes into consideration the square of error in forecasted 
# value over actual values. Thus, gives a much more accuracy.
# It is in our case is very low which teels that our model is accurately
# able to predict the values with existing variables 
# but must be considered in context of R square.
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor
```

```python
tree=DecisionTreeRegressor()
```

```python
treemodel=tree.fit(x,y)
```

```python
treemodel.score(x,y)
```

```python
tree2=DecisionTreeRegressor(max_depth=5)
```

```python
tree2model=tree2.fit(x,y)
```

```python
tree2model.score(x,y)
# With depth of 5 score is coming below 0.95.I have reduced the max depth to 5 to control the overfitting
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor
```

```python
RF=RandomForestRegressor(n_estimators=1000)
```

```python
RFmodel=RF.fit(x,y)
```

```python
RFmodel.score(x,y)
# The score is low compared to decision tree. it does not lead to overfitting 
# It considers various parameters as well as random subset of training data to form decision tree.
```

```python
# Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbm=GradientBoostingRegressor(n_estimators=3000)
```

```python
gbmodel=gbm.fit(x,y)
```

```python
gbmodel.score(x,y)
# Since, the result we get from Gradient Boosting Method is 1.0. 
# Thus, we can conveniently conclude that the method has accurately read the training data as it corrects the errors made by previous trees. But such
# high levels of accuracy can be a sign of overfitting.
```
